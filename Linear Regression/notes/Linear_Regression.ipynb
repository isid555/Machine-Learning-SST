```python
# Displaying our data
import pandas as pd
import numpy as np
data = pd.read_csv('/content/train-cars24-car-price.csv')
data.head()
```

```python
# using single feature to predict selling price of a car
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.title('MaxPower vs. Price')
sns.scatterplot(data=data, x='max_power', y='selling_price');
```

```python
def estimate_charges(maxpower, w, b):
    return w * maxpower + b
```

```python
w=50
b=100
```

```python
maxpower= data.max_power

estimated_charges = estimate_charges(maxpower, w, b)
```

```python
plt.figure(figsize=(10,6))
plt.plot(maxpower, estimated_charges, 'y-o');
plt.xlabel('MaxPower');
plt.ylabel('Estimated Price');
```

```python
target = data.selling_price
plt.figure(figsize=(10,6))
plt.plot(maxpower, estimated_charges, 'y', alpha=0.9);
plt.scatter(maxpower, target, s=8,alpha=0.8);
plt.xlabel('MaxPower');
plt.ylabel('Price')
plt.legend(['Estimate', 'Actual']);
```

```python
def try_parameters(w, b):
    maxpower = data.max_power
    target = data.selling_price

    estimated_charges = estimate_charges(maxpower, w, b)
    plt.figure(figsize=(10,6))
    plt.plot(maxpower, estimated_charges, 'r', alpha=0.9);
    plt.scatter(maxpower, target, s=8,alpha=0.8);
    plt.xlabel('MaxPower');
    plt.ylabel('Price')
    plt.legend(['Estimate', 'Actual']);
```

```python
try_parameters(0.5,1)
```

```python
try_parameters(0.08,0.5)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

!gdown 1UpLnYA48Vy_lGUMMLG-uQE1gf_Je12Lh
```

```python
df = pd.read_csv('cars24-car-price-clean.csv')
df.head()
```

```python
# define X and y
X = df["model"].values
Y = df["selling_price"].values
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
type(model)
```

```python
model.fit(X,Y)
```

```python
X.shape
```

```python
Y.shape
```

```python

X = X.reshape(X.size, 1)
Y = Y.reshape(Y.size, 1)

print(X.shape,Y.shape)
```

```python
model.fit(X,Y)
```

```python
y_hat = model.predict(X)

print('Predicted')
print(y_hat[:3])
print('Actual')
print(Y[:3])
```

```python
w0 = model.intercept_
w1 = model.coef_
print(w0, w1)
```

```python
fig = plt.figure()
plt.scatter(X,Y,label='data')
plt.scatter(X,y_hat,color='orange',label='prediction')
plt.legend()
plt.show()
```

```python
model.score(X,Y)
```

```python
# Statmodels implementation of Linear regression
import statsmodels.api as sm
# Lets scale the data, standardization
from sklearn.preprocessing import StandardScaler

X = df[df.columns.drop('selling_price')]
Y = df["selling_price"]

sc = StandardScaler()
cols = X.columns
X[cols] = sc.fit_transform(X[cols])
```

```python

X_sm = sm.add_constant(X)  #Statmodels default is without intercept, to add intercept we need to add constant

sm_model = sm.OLS(Y, X_sm).fit()

print(sm_model.summary())
```
